## Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set()
```

```python
df = pd.read_csv('netflix_titles_2021.csv')
df.head()
```

## Exploring Dataset

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.columns
```

```python
# Finding numeric columns
num_col=[]
for col in df.columns:
    if(df[col].dtypes != 'object'):
        num_col.append(col)
        
print(num_col)
```

```python
# Finding categorical columns
cat_col=[]
for col in df.columns:
    if(df[col].dtypes == 'object'):
        cat_col.append(col)
        
print(cat_col)
```

```python
df.isnull().sum()
```

```python
# Handling missing values
df['director'].fillna('Not available', inplace=True)
df['cast'].fillna('Not available', inplace=True)
df['country'].fillna('Not available', inplace=True)
df['date_added'].fillna(method='ffill', inplace=True)
df['rating'].fillna(method='ffill', inplace=True)
df['duration'].fillna(method='ffill', inplace=True)
```

```python
df.isnull().sum()
```

```python
df[['date_added', 'release_year']]
```

```python
# converting date_added to month and year
df['added_month'] = df['date_added'].apply(lambda x: x.split(',')[0].split()[0])
df['added_year'] = df['date_added'].apply(lambda x: x.split(',')[1])
df[['added_month', 'added_year']]
```

```python
df['added_year'] = df['added_year'].astype(int)
```

```python
df.drop('date_added', axis=1, inplace=True)
```

```python
df.describe()
```

```python
df.head()
```

```python
df.shape
```

```python
# date_added_year can't be less than release year
df.drop(df[df['added_year']<df['release_year']].index,inplace = True)
```

```python
df.shape
```
